## Regresión Logística

La regresión logística es un modelo de clasificación que utiliza una función logística para modelar la probabilidad de que una observación pertenezca a una clase particular.

### Objetivo
Predecir la probabilidad de que una observación pertenezca a una de dos categorías, basándose en las características disponibles.

### Conceptos Claves:
- **Función logística**: Es una función sigmoide que transforma valores reales en probabilidades, limitada entre 0 y 1.
  $$
  P(Y=1|X) = \frac{1}{1 + e^{-(\beta_0 + \beta_1X)}}
  $$
- **Coeficientes (β)**: Representan el impacto de cada característica en la probabilidad de la clase objetivo.
- **Log-odds**: La regresión logística modela los logaritmos de las probabilidades, lo cual permite una relación lineal entre las características y las probabilidades.
  $$
  \log\left(\frac{P}{1-P}\right) = \beta_0 + \beta_1X
  $$
- **Función de pérdida (entropía cruzada)**: Se utiliza para medir el error del modelo y optimizar los coeficientes.

### Variables requeridas
Pueden utilizarse variables continuas y categóricas (codificadas), y no es necesario que sigan una distribución normal.

### Algoritmo
1. **Preparación de los datos**: Codificar variables categóricas y estandarizar variables continuas si es necesario.
2. **Ajuste del modelo**: Utilizar el método de máxima verosimilitud para estimar los coeficientes.
3. **Predicción de probabilidades**: Calcular la probabilidad de que una observación pertenezca a la clase positiva.
4. **Clasificación**: Asignar una clase a partir de un umbral de probabilidad (generalmente 0.5).
5. **Ajuste del umbral** (opcional): Ajustar el umbral de probabilidad para optimizar métricas específicas.

### Evaluación
Se puede evaluar con métricas como precisión, exactitud, sensibilidad y especificidad, y mediante una matriz de confusión. También se utiliza el área bajo la curva ROC (AUC-ROC) para evaluar el rendimiento del modelo.
